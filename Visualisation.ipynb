```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv(r"C:\Users\gelek\OneDrive\Desktop\shootings.csv")
```

```python
# Check the data types and missing values
print(df.info())
```

```python
# Summary statistics
print(df.describe())
```

```python
# Distribution of outcome (death or injury)
plt.figure(figsize=(8,6))
sns.countplot(x='manner_of_death', data=df)
plt.title('Distribution of outcome')
plt.xlabel('Manner of death')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of age
plt.figure(figsize=(8,6))
sns.histplot(x='age', data=df, bins=30)
plt.title('Distribution of age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between age and threat level
plt.figure(figsize=(8,6))
sns.boxplot(x='threat_level', y='age', data=df)
plt.title('Relationship between age and threat level')
plt.xlabel('Threat level')
plt.ylabel('Age')
plt.show()
```

```python
# Frequency of shootings by state
plt.figure(figsize=(16,6))
sns.countplot(x='state', data=df, order=df['state'].value_counts().index)
plt.title('Frequency of shootings by state')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Racial breakdown of people shot
plt.figure(figsize=(8,6))
df['race'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Racial breakdown of people shot')
plt.ylabel('')
plt.show()
```

```python
# Relationship between race and outcome
plt.figure(figsize=(8,6))
sns.countplot(x='race', hue='manner_of_death', data=df)
plt.title('Relationship between race and outcome')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
```

This code loads the data, checks for data types and missing values, calculates summary statistics, and creates several visualizations using the matplotlib and seaborn libraries. The visualizations include a countplot to show the distribution of outcomes (death or injury), a histogram to show the distribution of age, a boxplot to show the relationship between age and threat level, a countplot to show the frequency of shootings by state, a pie chart to show the racial breakdown of people who were shot, and a countplot to show the relationship between race and outcome.
